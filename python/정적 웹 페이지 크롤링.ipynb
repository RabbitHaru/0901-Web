```python
from bs4 import BeautifulSoup
html = '<h1 id="title">한빛출판네트워크</h1><div class="top"><ul class="menu"><li><a href="http://www.hanbit.co.kr/member/login.html" class="login">로그인</a></li></ul><ul class="brand"><li><a href="http://www.hanbit.co.kr/member/media">한빛미디어</li><li><a href="http://www.hanbit.co.kr/member/academy">한빛아카데미</li></ul></div>'
soup = BeautifulSoup(html, 'html.parser')
print(soup)
print(soup.h1)
print(soup.div)
print(soup.a)
print(soup.ul)
print(soup.li)
# find_all("속성") : 모든 속성에 맞는 태그를 찾아 리스트로 반환
liTag = soup.find_all("a")
for i in liTag:
    print(i)
# find("속성") : 속성을 이용한 특정 태그를 찾아 반환
login = soup.find('a', attrs={'class':'login'})
print(login)
id = soup.find(attrs={'id':'title'})
print(id)
# select("css선택자") : css설정시 사용하는 선택자를 이용하여 찾는 함수
li_list = soup.select("div>ul.brand>li")
for li in li_list:
    print(li)
```

```python
from bs4 import BeautifulSoup
html = '<h1 id="title">한빛출판네트워크</h1><div class="top"><ul class="menu"><li><a href="http://www.hanbit.co.kr/member/login.html" class="login">로그인</a></li></ul><ul class="brand"><li><a href="http://www.hanbit.co.kr/member/media">한빛미디어</li><li><a href="http://www.hanbit.co.kr/member/academy">한빛아카데미</li></ul></div>'
soup = BeautifulSoup(html, 'html.parser')
aTag = soup.a
print(aTag)
# attrs : 속성들을 딕셔너리 형식으로 저장하는 변수
print(aTag.attrs)
# 딕셔너리에서 데이터를 꺼내는 방식으로 속성의 값을 출력할 수 있음
print(aTag['class'])
print(aTag['href'])
```

```python
# 필요한 라이브러리 추가
from bs4 import BeautifulSoup
import urllib.request
# 결과를 저장할 변수 선언
result = []
# 몇 페이지까지 데이터를 받아올지 설정하는 반복문
for page in range(1,46+1):
    # 데이터를 가지고올 페이지의 url 설정
    # pageNo가 1이면 1페이지, 2면 2페이지
    hollys_url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page}&sido=&gugun=&store='
    print(hollys_url)
    # url을 이용하여 실제 페이지의 html 코드를 변수 저장
    html = urllib.request.urlopen(hollys_url)
    # 각각의 html안의 태그를 검색할 수 있도록 파싱진행
    soupHollys = BeautifulSoup(html,'html.parser')
    # 매장정보가 들어있는 tbody 태그를 변수에 저장
    tag_tbody = soupHollys.find('tbody')
    # 열하나에 매장 하나가 들어있기 때문에 열을 기준으로 반복문 실행
    for store in tag_tbody.find_all("tr"):
        # tr태그의 문자열이 3글자 보다 작으면 반복문 종료
        if len(store) <= 3 :
            break
        # 각각의 데이터가 들어있는 td태그를 저장
        store_td = store.find_all('td')
        store_name = store_td[1].string # 매장명
        store_sido = store_td[0].string # 지역
        store_address = store_td[3].string # 주소
        store_phone = store_td[5].string # 전화번호
        store_flag = store_td[2].string # 현황
        # 매장정보를 list에 저장
        result.append([store_name]+[store_sido]+[store_address]+[store_phone])
for i in result:
    print(i)
import pandas as pd
# 리스트 자료형의 데이터를 판다스에서 사용하는 형식(테이블)으로 변경
hollys_tbl = pd.DataFrame(result, columns=('store','sido-gu','address','phone'))
# csv로 저장
# to_csv("파일경로", encoding="인코딩 형식", mode="파일모드r,w,a", index=번호출력여부)
hollys_tbl.to_csv("hollys.csv", encoding="cp949", mode="w", index=True)
```
