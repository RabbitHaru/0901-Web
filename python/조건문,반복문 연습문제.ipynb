```python
# 0 : 구구단 2~9단까지 출력하기 
a=2
while (a<=9):
    b=1
    while (b<=9):
        print(str(a) + "x"+ str(b) +"=" + str(a*b) )
        b+=1
    a+=1
    print()
```

```python
# 1 : while문을 사용해 1부터 1000까지의 자연수 중 3의 배수의 합을 구해 보자.
i=1
sum=0
while(i<=1000):
    if(i%3==0):
        sum+=i
    i+=1
print(sum)
```

```python
# 2 : a 리스트의 총합을 구해 보자.
a = [70,60,55,75,95,90,80,80,85,100]
i=0
sum=0
while (i<=9):
    sum+=a[i]
    i+=1
print(sum)
```

```python
# 3 : 1~100 중 홀수만을 저장하는 리스트를 만들어 보자
i=0
list=[]
while(i<=100):
    if(i%2==1):
        list.append(i)
    i+=1
print(list)
```

```python
# 4 : 리스트에서 글자 수가 세글자 이상인 문자만 출력해보자
a = ['Life','is','too','short',',','you','need','python']
i=0
while (i < len(a)):
    if  (len(a[i])>3):
        print(a[i])
    i+=1
```

```python
# 5 : 1900~2023년 까지의 윤년을 출력해보자
i=1900
while(i<=2300):
    if(i%4==0 and i%100!=0) or (i % 400 == 0):
        print(i)
    i+=1
```
